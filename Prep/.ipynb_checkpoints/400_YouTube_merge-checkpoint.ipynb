```python
import os.path
import pandas as pd
import numpy as np

import json#for decoding a JSON response
import pandas as pd
import seaborn as sns
import jellyfish
import numpy as np
from scipy import stats
#import plotly as py
#import cufflinks as cf
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
if  not os.path.exists("movies.csv"):
    print("Missing dataset file")
```

```python
movies = pd.read_csv('movies.csv',  encoding='latin-1')
```

```python
trailerData = pd.read_pickle('trailers.pkl')
```

```python
trailerData.head()
```

```python
trailerData.shape
```

```python
movieTrailers = movies.merge(trailerData, left_on='name', right_on='Title')
```

```python
movieTrailers
```

```python
movieTrailers = movieTrailers.sort_values('Views', ascending=False).drop_duplicates('Title',dropfirst=True).sort_index()
```

```python
movieTrailers
```

```python
movieTrailers.Published = movieTrailers.Published.astype(np.int64)
```

```python
movieTrailers = movieTrailers[(movieTrailers['year'] >= movieTrailers['Published'])]
```

```python
movieTrailers = movieTrailers.drop(['company','country','budget','name','director','star','writer','rating','released','runtime'], axis=1)
```

```python
movieTrailers
```

```python
movieTrailers.reset_index(inplace=True)
```

```python
movieTrailers.head()
```

```python
movieTrailers.Views = movieTrailers.Views.astype(np.int64)
movieTrailers.Likes = movieTrailers.Likes.astype(np.int64)
movieTrailers.Dislikes = movieTrailers.Dislikes.astype(np.int64)
movieTrailers.Comments = movieTrailers.Comments.astype(np.int64)
```

```python
movieTrailers.head()
```

```python
movieTrailers.to_pickle('movieTrailers.pkl')
```
