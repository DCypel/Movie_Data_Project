```python
import os.path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
from scipy import stats
%matplotlib inline
```

```python
oscarMovies = pd.read_pickle('oscarMovies.pkl')
```

# Oscar Winning movies

We see the effect that winning an oscar has on a movies success quite clearly here as there is a huge jump in the harmonic mean of oscar winning movies to that of movies without an oscar win. The movies that do not have an Oscar receive a harmonic mean socre of 0.4, we can see that this doubles for movie that have an Oscar win as they have an average harmonic mean of 0.8. The intesting effect here is that the differnce in a movies harmonic mean is much lower when we compare movies that have won oscars. In fact we see that the best number of oscar for a movie to have is two as the harmonic mean is best for those movies as it scores about a 0.85.

```python
oscarMovies.groupby('Oscar_wins')['HarMean'].mean().plot.bar(title ='Number of Oscar Wins',color='gold',fontsize=15,figsize=(10,5))
plt.ylabel('HarMean',size=18)
plt.xlabel('Number of Oscar Winners',size=18)
```

# Oscar Nominated movies

The results we find for oscar nominated movies are simialar as we see the largest jump in the harmonic mean was between movies with no nominations which had a harmonic socre of 0.4, and movie with an oscar nomination, whose harmoic score was 0.7. Whats interesting to note is that as the number of nominations for movies increased so did the harmonic mean of the movies. 

```python
oscarMovies.groupby('Oscar_noms')['HarMean'].mean().plot.bar(title='Number of Oscar Noms',color='gold',figsize=(10,5),fontsize=15)
plt.ylabel('Score',size=18)
plt.xlabel('Number of Oscar nominations',size=18)
```

# Gross and Score of Movies 

When we break down our analysis of the harmonic mean into gross and score we again found some interesting results.

```python
wins2 = oscarMovies[(oscarMovies['gross']<=350000000)&(oscarMovies['Oscar_wins']>=1)]#removes movies that grosses above 350,000,000 millionand movies witout an oscar win.
noms2 = oscarMovies[(oscarMovies['gross']<=350000000)&(oscarMovies['Oscar_noms']>=1)]
```

When we examined the Oscar nominated movies we found that tended to score qutie highyl grouping around a score of 7. We can also see that there was not much differnce in the score of movies that had been nominated a different number of times, as they are all randomly dispersed. When we examine the gross of these movies we can see that it appears to group on the low side of the gross, however this is a normal distribution for all movies as we see the same trend in movies without oscar nominations.

```python
noms2.plot.scatter('gross', 'score', c='Oscar_noms',s=40,title='Oscar Noms by Score and Gross', colormap='jet',sharex=False, figsize=(15,10),legend=True,fontsize=15,use_index=True,rot=0)
plt.ylabel('Score',size=15)
plt.xlabel('Gross (where 1 is 100 million)',size=15)
```

Taking a closer look at the average score of movies by the number of nominations they have we can see that there is very little difference in the socre between movies with one oscar nomination and movies with 4 oscar nominations.

```python
oscarMovies.groupby('Oscar_noms')['score'].mean().plot.bar(title='Average scores of movies by their oscars nominations',figsize=(10,8))
plt.ylabel('Score',size=18)
plt.xlabel('Number of Oscar nominations',size=18)
```

We see similar trend in our oscar winning movies as we did with our oscar nominated movies. The oscar winning movies score higher on average as we see a grouping between 7 and 8, giving us an average score of 7.5. Again there is no noticable difference between movies with a single oscar win and those with multiple wins in terms of the score of the movie. When we consider the gross of these movies we can see that they gross in a similar fashion as the oscar nominated movies, while it appears low this is in fact a normal distribution as the majority of movies gross beneath 100 million. 

```python
wins2.plot.scatter('gross', 'score', c='Oscar_wins',s=40, colormap='jet',sharex=False, figsize=(15,10),title='Movies by oscar wins compared to score and gross',legend=True,fontsize=15,use_index=True,rot=0)
plt.ylabel('Score',size=15)
plt.xlabel('Gross(where 1 is 100 million)',size=15)
```

Taking a closer look at the scorings of our oscar winning movies we can see that they all tend to come in around 7.5 with the only bar that is much lower is the bar that represents movies with no oscar wins. 

```python
oscarMovies.groupby('Oscar_wins')['score'].mean().plot.bar(title='Average scores of movies by their oscar wins',figsize=(10,8))
plt.ylabel('Score',size=18)
plt.xlabel('Number of Oscar Winners',size=18)
```
