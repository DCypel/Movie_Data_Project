```python
import os.path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Ensure the file exists
if  not os.path.exists("AllMoviesCastingRaw.csv"):
    print("Missing dataset file")
```

```python
casting = pd.read_csv("AllMoviesCastingRaw.csv", sep=';', engine = 'python')
```

```python
casting.head(5)
```

```python
casting.actor1_gender.replace([0, 1, 2], ['Unknown', 'Female', 'Male'], inplace=True)
casting.actor2_gender.replace([0, 1, 2], ['Unknown', 'Female', 'Male'], inplace=True)
casting.actor3_gender.replace([0, 1, 2], ['Unknown', 'Female', 'Male'], inplace=True)
casting.actor4_gender.replace([0, 1, 2], ['Unknown', 'Female', 'Male'], inplace=True)
casting.actor5_gender.replace([0, 1, 2], ['Unknown', 'Female', 'Male'], inplace=True)
casting.director_gender.replace([0, 1, 2], ['Unknown', 'Female', 'Male'], inplace=True)
```

```python
casting.head(5)
```

```python
casting.drop(['actor_number', 'director_number', 'producer_name', 'producer_number', 'screeplay_name', 'editor_name'
             ], axis=1, inplace=True)
```

```python
casting.head(5)
```

```python
casting.rename(columns={'id': 'ID','actor1_name': 'Actor 1', 'actor1_gender': 'Actor 1 Gender', 'actor2_name': 'Actor 2',
                        'actor2_gender': 'Actor 2 Gender', 'actor3_name': 'Actor 3', 'actor3_gender': 'Actor 3 Gender',
                        'actor4_name': 'Actor 4', 'actor4_gender': 'Actor 4 Gender', 'actor5_name': 'Actor 5', 
                        'actor5_gender': 'Actor 5 Gender','director_name': 'Director', 'director_gender': 'Director Gender'
                       }, inplace=True)
```
