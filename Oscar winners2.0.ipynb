```python
import os.path
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import io
%matplotlib inline
```

```python
awards = pd.read_csv('academy_awards.csv',  encoding='latin-1')
```

```python
awards.head()
```

```python
awards.drop(['Unnamed: 5', 'Unnamed: 6','Unnamed: 7','Unnamed: 8','Unnamed: 9','Unnamed: 10',], axis=1, inplace=True)
```

```python
awards.rename(columns={'Won?': 'Winner','Additional Info': 'Movie'}, inplace=True)
```

```python
nominations =  awards[awards.Winner == 'NO']
```

```python
awards =  awards[awards.Winner == 'YES']
```

```python
awards.head()
```

```python
awards.shape
```

```python
nominations.shape
```

```python
Directing =  awards[awards.Category == 'Directing']
leadActor =  awards[awards.Category == 'Actor -- Leading Role']
leadActress =  awards[awards.Category == 'Actress -- Leading Role']
supportActor =  awards[awards.Category == 'Actor -- Supporting Role']
supportActress =  awards[awards.Category == 'Actress -- Supporting Role']

DirectingNom =  nominations[nominations.Category == 'Directing']
leadActorNom =  nominations[nominations.Category == 'Actor -- Leading Role']
leadActressNom =  nominations[nominations.Category == 'Actress -- Leading Role']
supportActorNom =  nominations[nominations.Category == 'Actor -- Supporting Role']
supportActressNom =  nominations[nominations.Category == 'Actress -- Supporting Role']
```

```python
leadActorNom.head()
```

```python
leadActress.shape
```

```python
leadActor.head()
```

```python
oscar_win = pd.concat([Directing,leadActor,leadActress,supportActor,supportActress],axis=0,ignore_index=True)

oscar_nom = pd.concat([DirectingNom,leadActorNom,leadActressNom,supportActorNom,supportActressNom],axis=0,ignore_index=True)
```

```python
Director_idx = (oscar_win['Category'] == 'Directing')
oscar_win.loc[Director_idx,['Nominee','Movie']] = oscar_win.loc[Director_idx,['Movie','Nominee']].values

DirectorNom_idx = (oscar_nom['Category'] == 'Directing')
oscar_nom.loc[DirectorNom_idx,['Nominee','Movie']] = oscar_nom.loc[DirectorNom_idx,['Movie','Nominee']].values
```

```python
oscar_nom.groupby('Category').head()
```

```python
oscar_win = oscar_win.replace(dict(Movie={'\ {.*\}': ''}), regex=True)
oscar_win = oscar_win.replace(dict(Year={'\(.*\)': ''}), regex=True)

oscar_nom = oscar_nom.replace(dict(Movie={'\ {.*\}': ''}), regex=True)
oscar_nom = oscar_nom.replace(dict(Year={'\(.*\)': ''}), regex=True)
```

```python
oscar_win = oscar_win.replace(dict(Nominee={'\ and.*': ''}), regex=True)

oscar_nom = oscar_nom.replace(dict(Nominee={'\ and.*': ''}), regex=True)
```
