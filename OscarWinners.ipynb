```python
awards =  awards[awards.Award != 'Outstanding Picture']
```

```python
actor =  awards[awards.Award == 'Actor']
actress =  awards[awards.Award == 'Actress']
Directing =  awards[awards.Award == 'Directing']
AssDirector = awards[awards.Award == 'Assistant Director']
leadActor =  awards[awards.Award == 'Actor in a Leading Role']
leadActress =  awards[awards.Award == 'Actress in a Leading Role']
supportActor =  awards[awards.Award == 'Actor in a Supporting Role']
supportActress =  awards[awards.Award == 'Actress in a Supporting Role']
editing =  awards[awards.Award == 'Film Editing']
Writing1 =  awards[awards.Award == 'Writing (Adapted Screenplay)']
Writing2 =  awards[awards.Award == 'Writing (Original Screenplay)']
Writing3 =  awards[awards.Award == 'Writing (Screenplay Adapted from Other Material)']
Writing4 =  awards[awards.Award == 'Writing (Screenplay Based on Material Previously Produced or Published)']
Writing5 =  awards[awards.Award == 'Writing (Screenplay Based on Material from Another Medium)']
Writing6 =  awards[awards.Award == 'Writing (Screenplay Written Directly for the Screen)']
Writing7 =  awards[awards.Award == 'Writing (Screenplay)']
Writing8 =  awards[awards.Award == 'Writing (Screenplay, Based on Material from Another Medium)']
Writing9 =  awards[awards.Award == 'Writing (Screenplay, Original)']
Writing10 =  awards[awards.Award == 'Writing (Story and Screenplay)']
Writing11 =  awards[awards.Award == 'Writing (Story and Screenplay, Written Directly for the Screen)']
Writing12 =  awards[awards.Award == 'Writing (Story and Screenplay, Based on Material Not Previously Published or Produced)']
Writing13 = awards[awards.Award == 'Writing (Story and Screenplay, Based on Factual Material or Material Not Previously Published or Produced)']
```
